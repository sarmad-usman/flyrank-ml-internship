## 3. Human review + the no-go list

*What a person must check before acting. What should never be automated.*

```python
human_review_and_no_go = {
    "human_review_checklist": [
        "Verify that the content in question truly violates policy (e.g., manual review of flagged content).",
        "Confirm the severity of the violation aligns with the proposed action (e.g., temporary suspension vs. permanent ban).",
        "Assess potential cultural or linguistic nuances that automated systems might miss.",
        "Review the historical context of the user/content creator before severe actions are taken.",
        "Ensure any promotional action aligns with brand safety guidelines and current marketing strategies."
    ],
    "no_go_list_for_automation": [
        "Permanent account deletions or irreversible content removal without human confirmation.",
        "Legal notices or direct communication with law enforcement.",
        "Any action that could result in significant financial loss or legal liability for the platform or user.",
        "Decisions requiring subjective interpretation of complex ethical guidelines.",
        "Actions impacting sensitive user data or privacy without explicit consent and human oversight."
    ]
}

print("Human Review Checklist:")
for item in human_review_and_no_go['human_review_checklist']:
    print(f"  - {item}")

print("\nNo-Go List for Automation:")
for item in human_review_and_no_go['no_go_list_for_automation']:
    print(f"  - {item}")
```

## 4. Monitoring / retrain triggers

*What would tell you the recommendations went stale?*

```python
monitoring_retrain_triggers = {
    "monitoring_metrics": [
        {
            "metric_name": "Action Recommendation Acceptance Rate",
            "description": "Percentage of recommended actions that are actually taken by human reviewers.",
            "threshold_for_alert": "< 70% consistently for 3 days"
        },
        {
            "metric_name": "Content Flag Resolution Time",
            "description": "Average time taken to resolve content flags after an action is recommended.",
            "threshold_for_alert": "> 24 hours increase for a week"
        },
        {
            "metric_name": "False Positive Rate of Violations",
            "description": "Rate at which recommended 'policy violation' actions are overturned by human review.",
            "threshold_for_alert": "> 15% for a week"
        },
        {
            "metric_name": "Engagement Metric Shift",
            "description": "Significant unexpected changes in overall user engagement metrics (e.g., watch time, shares) following actions.",
            "threshold_for_alert": "5% absolute drop or rise for a week"
        }
    ],
    "retrain_triggers": [
        "Significant shift in underlying data distributions (e.g., new content types, user demographics).",
        "Deployment of major product changes affecting user interaction or content submission.",
        "Updates to platform policy or community guidelines.",
        "Performance degradation observed through monitoring metrics exceeding predefined thresholds.",
        "Annual review cycle, regardless of performance metrics."
    ]
}

print("Monitoring Metrics and Retrain Triggers:")
print("\nMonitoring Metrics:")
for metric in monitoring_retrain_triggers['monitoring_metrics']:
    print(f"  - {metric['metric_name']}: {metric['description']} (Alert if: {metric['threshold_for_alert']})")

print("\nRetrain Triggers:")
for trigger in monitoring_retrain_triggers['retrain_triggers']:
    print(f"  - {trigger}")
```
